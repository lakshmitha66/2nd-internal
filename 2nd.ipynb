```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_excel('demo.xlsx')
data
```

```python
data.info()
```

## 1.Checking for missing values

```python
data.isnull()
```

## 2.	Dropping the row where there are missing values

```python
data.dropna(axis=1,inplace=True)
data
```

## 3.	Replacing with mean for the missing value 

```python
data.fillna(0)
```

```python
data['age'].fillna(data['age'].mean(),inplace=True)
data
```

```python
data['date'].fillna(data['date'].mean(),inplace=True)
data
```

## 4.	Replacing with Mode for the missing value 

```python
data['gender'].fillna(data['gender'].mode(),inplace=True)
data
```

## 5.	Find and remove the outlier 

```python
plt.boxplot(data['age'])
```

```python
for i in data['age']:
    q1=data['age'].quantile(0.25)
    q3=data['age'].quantile(0.75)
    iqr=q3-q1
    lower_tail=q1-1.5*iqr
    upper_tail=q3+1.5*iqr
    if i>upper_tail or i< lower_tail:
        data['age']=data['age'].replace(i,np.mean(data['age']))
plt.boxplot(data['age'])
plt.show()
```

```python
for i in data['age']:
    q1=data['age'].quantile(0.25)
    q3=data['age'].quantile(0.75)
    iqr=q3-q1
    lower_tail=q1-1.5*iqr
    upper_tail=q3+1.5*iqr
    if i>upper_tail or i< lower_tail:
        data['age']=data['age'].drop(i)
plt.boxplot(data['age'])
plt.show()
```

## 6.	Normalize the continuous numerical features.


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(data[['age','date']]) 
data_normalized = scaler.transform(data[['age','date']]) 
data_normalized
```

```python
data_normalized=pd.DataFrame(data_normalized,columns=['age','date'])
data_normalized=data_normalized.join(data[data.columns.drop(['age','date'])])
data_normalized
```

```python
Y = data_normalized['age']
X = data_normalized[data_normalized.columns.drop('age')]
print(X.shape)
```
